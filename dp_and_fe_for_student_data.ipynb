# ทำให้ชื่อคณะเป็นมาตรฐานเดียวกัน

```python
def standardize_faculty(คณะ):
    if คณะ in ['วิทยาศาสตร์', 'วิทย์', 'Science', 'คณะวิทยาศาสตร์','Sci']:
        return 'วิทยาศาสตร์'
    elif คณะ in['ศิลปศาสตร์', 'ศิลป์', 'Arts', 'Liberal Arts']:
        return 'ศิลปศาสตร์'
    elif คณะ in['วิศวกรรมศาสตร์', 'วิศวฯ', 'วิศวกรรม', 'วิศวะ', 'Engineering']:
        return 'วิศวกรรมศาสตร์'
    else:
        return คณะ
```

```python
df['คณะ'] = df['คณะ'].apply(standardize_faculty)
```

```python
df
```

# ทำให้เพศเป็นมาตรฐานเดียวกัน

```python
def standardize_gender(เพศ):
    if เพศ in ['ชาย', 'ช', 'M', 'Male']:
        return 'ชาย'
    elif เพศ in['หญิง', 'ญ', 'F',  'Female']:
        return 'หญิง'
    else:
        return เพศ
```

```python
df['เพศ']  = df['เพศ'].apply(standardize_gender)
```

```python
df
```

# แทนค่าอายุของคนที่ผิดปกติด้วยค่าเฉลี่ย

```python
# วิธีง่ายๆ - ใช้ IQR (Interquartile Range)
Q1 = df['อายุ'].quantile(0.25)
Q3 = df['อายุ'].quantile(0.75)
IQR = Q3 - Q1

# กำหนดขอบเขต
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_mean_age = df[(df['อายุ'] >= lower_bound) & (df['อายุ'] <= upper_bound)] ['อายุ'].mean()

df.loc[(df['อายุ'] < lower_bound) | (df['อายุ'] > upper_bound), 'อายุ' ] = int (round(df_mean_age))
```

```python
df
```

# แทนค่าเกรดของคนที่ผิดปกติด้วยค่าเฉลี่ย

```python
# วิธีง่ายๆ - ใช้ IQR (Interquartile Range)
Q1 = df['GPA'].quantile(0.25)
Q3 = df['GPA'].quantile(0.75)
IQR = Q3 - Q1

# กำหนดขอบเขต
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_mean_gpa = df[(df['GPA'] >= lower_bound) & (df['GPA'] <= upper_bound)] ['GPA'].mean()

df.loc[(df['GPA'] < lower_bound) | (df['GPA'] > upper_bound), 'GPA' ] = int (round(df_mean_gpa, 2))
```
